```python
import pandas as pd
import random
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
# Create a DataFrame for the 4 'bins'. For this purpose, I've used random numbers to fill in the dataframe.
d = {'SI':[30,33,21,33,42,52],'SA':[19,28,27,30,34,32],'PD':[49,40,28,46,30,43],'LS':[64,56,40,41,48,73]}
years = ['2016','2017','2018','2019','2020','2021']
df= pd.DataFrame(data=d, index=years)
totals = [i+j+k+l for i,j,k,l in zip(df['SI'], df['SA'], df['PD'], df['LS'])]
SI_l = [i / j * 100 for i,j in zip(df['SI'], totals)]
SA_l = [i / j * 100 for i,j in zip(df['SA'], totals)]
PD_l = [i / j * 100 for i,j in zip(df['PD'], totals)]
LS_l = [i / j * 100 for i,j in zip(df['LS'], totals)]
d_scale = {'SI':SI_l,'SA':SA_l,'PD':PD_l,'LS':LS_l}
df_scale = pd.DataFrame(data=d_scale, index=years)
```

```python
# scaled line plot
plt.plot(df_SI,color='blue',marker='o', markerfacecolor='blue', markersize=4)
plt.plot(df_SA,color='orange', marker='o', markerfacecolor='orange', markersize=4)
plt.plot(df_PD,color='green', marker='o', markerfacecolor='green', markersize=4)
plt.plot(df_LS,color='red', marker='o', markerfacecolor='red', markersize=4)
plt.legend(['Self Improvement (SI)','Social Advice (SA)','Professional Development (PD)','Learning Strategies (LS)'],bbox_to_anchor=(1.01,1.05))
plt.grid(True, color='k', linestyle=':')
plt.title("Label Trends from 2016 to 2021")
plt.xlabel("Year")
plt.ylabel("Percentage (%)")
plt.show()
```

```python
# SI Line plot
plt.plot(df_SI,color='blue',marker='o', markerfacecolor='blue', markersize=4)
rcParams['figure.figsize'] = 10, 4
plt.plot(df_SI,color='blue')
plt.grid(True, color='k', linestyle=':')
plt.title("Self Improvement Trends from 2016 to 2021")
plt.xlabel("Year")
plt.ylabel("Percentage (%)")
plt.ylim(10,40)
plt.show()
```
